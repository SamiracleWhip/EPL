```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Create 2 new columns
#minutes per match:
df["MinutesperMatch"]= (df["Mins"]/df["Matches"]).astype(int)

#goals per match
df["GoalsperMatch"]=(df["Goals"]/df["Matches"]).astype(float)
```

```python
df.head()
```

```python
#Calculate total goals
Total_Goals = df["Goals"].sum()
```

```python
Total_Goals
```

```python
#Out of these 986 goals, how many were scored by penalties?

Total_PenaltyGoals = df["Penalty_Goals"].sum()
print(Total_PenaltyGoals)
```

```python
Total_PenaltyGoals/Total_Goals
```

```python
#Penalty Attempts vs. Converted

Total_PenaltyAttempts = df["Penalty_Attempted"].sum()
```

```python
Total_PenaltyAttempts
```

```python
Total_PenaltyGoals/Total_PenaltyAttempts
```

```python
#Create a pie chart for penalties missed vs. scored

plt.figure(figsize=(13,6))
pl_not_scored = df["Penalty_Attempted"].sum()- Total_PenaltyGoals
data=[pl_not_scored,Total_PenaltyGoals]
labels=["Penalties missed","Penalties Scored"]
color = sns.color_palette("Set2")
plt.pie(data,labels = labels,colors=color,autopct="%.0f%%")
plt.show()
```

```python
#Most players come from which nations?
nationality = df.groupby("Nationality").size().sort_values(ascending=False)
nationality.head(10).plot(kind="bar",figsize=(12,4),color=sns.color_palette("magma"))
```

```python
#The clubs with maximum players in the squad
df["Club"].value_counts().nlargest(5).plot(kind="bar",color=sns.color_palette("viridis"))
```

```python
#Check the clubs with the least players in the squad

df["Club"].value_counts().nsmallest(5).plot(kind="bar",color=sns.color_palette("viridis"))
```

```python
#Players based on Age Groups. We will define the age groups

Under20 = df[df["Age"]<=20]
Age20_25 = df[(df["Age"]>20) & (df["Age"] <= 25)]
Age25_30 = df[(df["Age"]>25) & (df["Age"] <= 30)]
Above30 = df[df["Age"]>30]
```

```python
#create a pie chart to show this:
x= np.array([Under20["Name"].count(),Age20_25["Name"].count(),Age25_30["Name"].count(),Above30["Name"].count()])
mylabels= ["<=20",">20 & <=25","25 & <= 30",">30"]
plt.title("Total Players with Age",fontsize=10)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
#Total under 20 players in each club

players_under_20 = df[df["Age"]<20]
players_under_20["Club"].value_counts().plot(kind="bar",color=sns.color_palette("cubehelix"))
```

```python
#Under 20 players in ManU

players_under_20[players_under_20["Club"]=="Manchester United"]
```

```python
#Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x = "Club",y="Age",data=df)
plt.xticks(rotation=90)
```

```python
#Calculate Avg. Age per Club
num_player = df.groupby("Club").size()
data = (df.groupby("Club")["Age"].sum())/num_player
data.sort_values(ascending=False)
```

```python
#Total Assists from each club

Assists_by_clubs = pd.DataFrame(df.groupby("Club",as_index=False)["Assists"].sum())
```

```python
Assists_by_clubs.head()
```

```python
#Plot total Assists from each club

sns.set(style="whitegrid",color_codes=True)
ax = sns.barplot(x="Club",y="Assists",data=Assists_by_clubs.sort_values(by="Assists"),palette ="Set2")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title("Plot of Clubs vs Total Assists ",fontsize=20)
```

```python
#Find top 10 players with assists by the club and order by assist numbers
top_10_assists = df[["Name","Club","Assists","Matches"]].nlargest(n=10,columns="Assists")
top_10_assists
```

```python
#Plot Goals by clubs
Goals_by_clubs = pd.DataFrame(df.groupby("Club",as_index=False)["Goals"].sum())
```

```python
sns.set(style="whitegrid",color_codes=True)
ax = sns.barplot(x="Club",y="Goals",data=Goals_by_clubs.sort_values(by="Goals"),palette ="rocket")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,10)
plt.title("Plot of Clubs vs Total Goals ",fontsize=20)
```

```python
#Most goals by players

top_10_goals = df[["Name","Club","Goals","Matches"]].nlargest(n=10,columns="Goals")
top_10_goals
```

```python
#Create a pie chart of goals with assists and without assists

plt.figure(figsize=(14,7))
assists = df["Assists"].sum()
data=[Total_Goals - assists, assists]
labels = ["Goals without assists","Goals with assists"]
color = sns.color_palette("Set1")
plt.pie(data,labels=labels,colors=color,autopct ="%.0f%%")
plt.show()
```
